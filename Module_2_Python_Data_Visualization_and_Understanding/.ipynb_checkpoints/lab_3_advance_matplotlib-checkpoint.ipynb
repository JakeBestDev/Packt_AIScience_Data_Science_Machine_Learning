# Advanced Matplotlib

## Intro To Matplotlib Styles

### Matplotlib is built on top of pyplot
### Primarily used for data visualizations which is what a majority of this notebook is going to revolve around

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## First lets see how many different styles there are for visualizations and then pick for our visualizations

plt.style.available
```

```python
## Lets try the classic style
## We'll use the basic structure of a plot, without including any data just to show the style off

plt.style.use('classic')
fig = plt.figure()
ax = plt.axes()
```

```python
## Now lets try another style, 'dark_background'

plt.style.use('dark_background')
fig = plt.figure()
ax = plt.axes()
```

```python
## Now lets try another style, 'seaborn-v0_8-whitegrid'

plt.style.use('seaborn-v0_8-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
## Now lets try another style, 'ggplot'

plt.style.use('ggplot')
fig = plt.figure()
ax = plt.axes()
```
