## Matplotlib: Multiple Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## Lets first generate a new plot and add some data


x = np.linspace(0,10,1000) ## Lets generate 1000 points from 0-10
y = np.sin(x) ## Let us now use np.sin() and input our x values

plt.style.use('fivethirtyeight') ## Lets try a new style
fig = plt.figure() ## Create a new figure, or activate an existing figure.
ax=plt.axes() ## Add an Axes to the current figure and make it the current Axes.

ax.plot(x,y) ## Now lets plot x and y on our figure ax
```

```python
## We can all create a similar plot like above using just the following:
## Notice how it saves the style 
## We use ax to save the Axes

plt.plot(x,y)
```

```python
## Now lets create a y1, y2 and y3 using np.cos and the formula y+y2-0.5
## y1 = y = np.sin(x)
## y2 = np.cos(x)
## y3 = y+y2-0.5

y1 = np.sin(x)
y2 = np.cos(x)
y3 = y+y2-0.5
```

```python
## Now lets plot y1,y2,y3
## Notice how we now have 3 different curves 
## Notice how it is difficult to tell which plot is which. We don't have a legend. We will go into detail on that in a later lesson


plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
## We can take the above and easily change the styles
## The style will be carried on through the whole notebook unless changed

plt.style.use('bmh')
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
## And again we can do the same thing using our ax variable and build a figure like above instead of using plt.plot()

fig = plt.figure()
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

## Matplotlib: Line Colors and Styles

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## You are able to set colors and line styles within the plt.plot() statement
## First lets set our x and y values
## We will have y1,y2,y3 

x = np.linspace(1,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## Now lets set up the figure

plt.style.use('_mpl-gallery') ## Lets use a new style
fig = plt.figure(figsize=(4,2)) ## Set the size of the figure 
ax = plt.axes() ## Lets establish the axes

ax.plot(x,y1,color='blue') ## Lets plot y1 with color blue
ax.plot(x,y2,color='g') ## Lets plot y2 with color 'g' or green to show a different way of calling a color
ax.plot(x,y3,color=(0.8,0.6,0.7)) ## Lets plot y3 with a specific color by defining the RGB color scales
```

```python
## Not only the colors we can set the linestyles

plt.style.use('seaborn-v0_8-bright') ## Lets use a new style
fig = plt.figure(figsize=(4,2)) ## Set the size of the figure 
ax = plt.axes() ## Lets establish the axes

ax.plot(x,y1,color='blue', linestyle='solid') ## Lets plot y1 with a solid line style
ax.plot(x,y2,color='g',linestyle='dashed') ## Lets plot y2 with a dashed line style
ax.plot(x,y3,color=(0.8,0.6,0.7),linestyle='dashdot') ## Lets plot y3 with a dotted dash line style
```

## Matplotlib: Line Color and Style Shortcuts

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## Lets define some variables for our plots
## Lets use the same x, y1,y2,y3 we have been using
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## Lets first show our original plots from eailier and how the line styles and colors we're defined

plt.style.use('seaborn-v0_8-bright') ## Lets use a new style
fig = plt.figure(figsize=(4,2)) ## Set the size of the figure 
ax = plt.axes() ## Lets establish the axes

ax.plot(x,y1,color='blue', linestyle='solid') ## Lets plot y1 with a solid line style
ax.plot(x,y2,color='g',linestyle='dashed') ## Lets plot y2 with a dashed line style
ax.plot(x,y3,color=(0.8,0.6,0.7),linestyle='dashdot') ## Lets plot y3 with a dotted dash line style
```

```python
## We can use shortcuts to define colors and types of lines on a plot
## Notice how we don't define the line style or color, we instead use shorthand like:
## '-b' which means a solid blue line like in y1

plt.style.use('fivethirtyeight') ## Lets set a new style
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.plot(x,y1,'-b') ## '-b' which means a solid blue line
ax.plot(x,y2,'--g') ## '--g' which means a dashed green line
ax.plot(x,y3,':r') ## ':r' which means dotted red line
```

```python
## Now that we can see the shorthand versions of these plots lets try using linewidth to change the size of the lines

plt.style.use('fast') ## Lets set a new style
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.plot(x,y1,'-b',linewidth=8) ## Lets set blue to have a thicker line with linewidth
ax.plot(x,y2,'--g') 
ax.plot(x,y3,':r') 
```

```python
## We can use the '??' after plot to see all the shorthand and parameters within the plot function
## Look at the characters and linestyles to see more shorthands for formats
ax.plot??
```

## Matplotlib: Axis Limits

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## First lets define our x and y1,y2,y3
## We'll use the same ones as we have been

x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## We can define axis limits the better view the data presented in a plot
## The axes and Axis is a little different where:
## Axes will be to the confines of the plot whereas
## Axis can be defined outside the plot confines
## Both of which can be used in tandem
## See the examples below

## Lets first show an axes plot

plt.style.use('seaborn-v0_8-pastel') ## Lets try a new style
plt.figure(figsize=(4,2),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r',linewidth=0.8) ## Lets create a red dash dot plot with a line width of 0.8
```

```python
## So now that we have our figure defined lets try defining some axes limits
## We can limit the x boundries using xlim() while also usinge axes function

plt.figure(figsize=(4,2),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r',linewidth=0.8) ## Lets create a red dash dot plot with a line width of 0.8

plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
```

```python
## We can do the same thing with y using ylim()

plt.figure(figsize=(4,2),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r',linewidth=0.8) ## Lets create a red dash dot plot with a line width of 0.8

plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
plt.ylim(-1.5,1.5) ## So now we see that the limit of y is now set from -1.5 to 1.5
```

```python
## Lets add additional plots and see what it looks like with these limits
## Notice how the black line goes outside the boundries of the plot view with the ylim()

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
plt.ylim(-1.5,1.5) ## So now we see that the limit of y is now set from -1.5 to 1.5
```

```python
## If we remove the limits we can see it fits the plots entirley 

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

#plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
#plt.ylim(-1.5,1.5) ## So now we see that the limit of y is now set from -1.5 to 1.5
```

```python
## Resetting our plot 

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
plt.ylim(-1.5,1.5) ## So now we see that the limit of y is now set from -1.5 to 1.5
```

```python
## Instead of using the xlim() and ylim() functions we can use the axis() function to define our plot view
## Notice how the plots are the same with less lines of code thanks to the axis

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.axis([-2,12,-1.5,1.5])
```

```python
## The cool thing about the axis function is we can reverse the functions
## Lets try just reversing the x axis 2,12 to 12,2
## Notice how the plot has its x axis flipped

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.axis([12,-2,-1.5,1.5])
```

```python
## The cool thing about the axis function is we can reverse the functions
## Lets try just reversing the y axis to 1.5,-1.5 and changing the x axis back to -2,12
## Notice how the plot has its y axis has been flipped and the graph has totally flipped over

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.axis([-2,12,1.5,-1.5])
```

```python
## The cool thing about the axis function is we can reverse the functions
## Lets flip both the x and y axis

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.axis([12,-2,1.5,-1.5])
```

```python
## One last thing we can do with the axis function is use keywords
## So for this example we will use 'tight' to have the plot view tighten around the plot

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.axis('tight')
```

```python
## For this example we will use 'equal' to have the plot view tighten around the plot
## Notice how the aspect ratio changes

plt.figure(figsize=(8,4),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r') ## Lets create a red dash dot plot
ax.plot(x,y2,':b') ## Lets create a dotted blue line
ax.plot(x,y3,'-k') ## Lets create a solid black line

plt.axis('equal')
```

## Matplotlib: Legends Labels

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## First lets set up some data
## We'll use the same data as we have been

x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## For a figure we'd like to annotate we can add things like legends to label plots
## Lets create the same graph we used in our axis notes
## Notice how the legend is mixed into the plot though. That is an issue. See next example to set legend location

plt.style.use('seaborn-v0_8-ticks') ## Lets use a new style
fig = plt.figure(figsize=(4,2),dpi=100) ## set dpi to 100 and fig size to be 4,2
ax = plt.axes()
ax.plot(x,y1,'--b',label='sin(x)') ## Blue dashed line, add the label
ax.plot(x,y2,'-.g',label='cos(x)') ## green dash dotted line
ax.plot(x,y3,':r',label='sin(x) + cos(x) - 0.5') ## red dotted line
plt.axis('equal') ## set axis to be equal so we can see the whole plot
plt.legend()
```

```python
## We can change the location of the legend to be in different areas of the plot. 
## However notice that the moved legend is still in an awkward location

plt.style.use('seaborn-v0_8-ticks') ## Lets use a new style
fig = plt.figure(figsize=(4,2),dpi=100) ## set dpi to 100 and fig size to be 4,2
ax = plt.axes()
ax.plot(x,y1,'--b',label='sin(x)') ## Blue dashed line, add the label
ax.plot(x,y2,'-.g',label='cos(x)') ## green dash dotted line
ax.plot(x,y3,':r',label='sin(x) + cos(x) - 0.5') ## red dotted line
plt.axis('equal') ## set axis to be equal so we can see the whole plot

plt.legend(loc='upper left')
```

```python
## The lower left seems to be better but to get the best fix for the plot you will wanna increase the figsize

plt.style.use('seaborn-v0_8-ticks') ## Lets use a new style
fig = plt.figure(figsize=(4,2),dpi=100) ## set dpi to 100 and fig size to be 4,2
ax = plt.axes()
ax.plot(x,y1,'--b',label='sin(x)') ## Blue dashed line, add the label
ax.plot(x,y2,'-.g',label='cos(x)') ## green dash dotted line
ax.plot(x,y3,':r',label='sin(x) + cos(x) - 0.5') ## red dotted line
plt.axis('equal') ## set axis to be equal so we can see the whole plot

plt.legend(loc='lower left')
```

```python
## We can also set the title of a figure

plt.style.use('seaborn-v0_8-ticks') ## Lets use a new style
fig = plt.figure(figsize=(4,2),dpi=100) ## set dpi to 100 and fig size to be 4,2
ax = plt.axes()
ax.plot(x,y1,'--b',label='sin(x)') ## Blue dashed line, add the label
ax.plot(x,y2,'-.g',label='cos(x)') ## green dash dotted line
ax.plot(x,y3,':r',label='sin(x) + cos(x) - 0.5') ## red dotted line
plt.axis('equal') ## set axis to be equal so we can see the whole plot
plt.legend(loc='lower left')

plt.title('Sin-Cos-Curves')
```

```python
## We can set the x label and the y label

plt.style.use('seaborn-v0_8-ticks') ## Lets use a new style
fig = plt.figure(figsize=(4,2),dpi=100) ## set dpi to 100 and fig size to be 4,2
ax = plt.axes()
ax.plot(x,y1,'--b',label='sin(x)') ## Blue dashed line, add the label
ax.plot(x,y2,'-.g',label='cos(x)') ## green dash dotted line
ax.plot(x,y3,':r',label='sin(x) + cos(x) - 0.5') ## red dotted line
plt.axis('equal') ## set axis to be equal so we can see the whole plot
plt.legend(loc='lower left')
plt.title('Sin-Cos-Curves')

plt.xlabel('X Values')
plt.ylabel('Y Values')
```
