```python
## Folium is a library that allows you to analyze location data like latitude and logitude!
## Lets first import folumn
!pip install folium
import folium
```

```python
## Lets create a map in folium

m = folium.Map()
m
```

```python
## Lets import the covid-19 Dataset and assign it to a dataframe df using pandas
import pandas as pd

df = pd.read_csv('covid_19_till_19_May_2020.csv')
df.head()
```

```python
## Now lets create a folium map based on the above Covid Data.
## Lets use the lat and long to determine the location and then using the Confirmed column lets have a circle grow in size based on the number of confirmed cases

## First create a folium map

m = folium.Map()

## Now lets set the date column to datetime as this data frame contains the same countries at different times with multiple confirmed cases of Covid 
df['Date'] = pd.to_datetime(df['Date'])

## Now lets filter for the latest date in the dataframe which would have the most cases 

df2 = df[df['Date'] == max(df['Date'])]
```

```python
## Now you can see that this df2 has all of the most up to date confirmed cases for each country

df2
```

```python
## Now that we have the data we will create a loop of the length of data and plot using folium

## So the loop may look a little complicated but it isn't! 
## It is a very similar structure that we have used for previous graphs! 
## The main difference is that we are using a loop to plot each row and then adding it to graph m
## We are using iloc to located in the dataframe where that particular Lat,Long and Confirmed value is and then running the loop again

m = folium.Map()

for i in range(0,len(df2)):
    
    folium.Circle(location=[df2.iloc[i]['Lat'],df2.iloc[i]['Long']],
                 color = 'crimson', fill = 'crimson',
                 radius=int(df2.iloc[i]['Confirmed'])).add_to(m)

m
```

## Folium: Covid-19 Quiz
#### Create a Folium Map using the Covid-19 Dataframe
#### Use the minnimun date in place of the max date
#### Set the radius to be with respect to the number of deaths

```python
## First lets import folium, pandas and the Covid data

import folium
import pandas as pd

df = pd.read_csv('covid_19_till_19_May_2020.csv')
df.head()
```

```python
## Now lets convert the Date field to a datetime field

df['Date'] = pd.to_datetime(df['Date'])
```

```python
## Good now its a datetime field 

df.info()
```

```python
## Now that the date is a datetime field lets filter the min dates

df2 = df[df['Date']==min(df['Date'])]
df2.head()
```

```python
## Now lets create our folium map and plot using a for loop

m = folium.Map()

for i in range(0,len(df2)):
    folium.Circle(location=[df2.iloc[i]['Lat'],df2.iloc[i]['Long']],
                 color='crimson',fill='crimson',
                 radius=int(df2.iloc[i]['Deaths'])).add_to(m)

m
```
