## Intro to seaborn

```python
## Previous libraries
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## This is the intro to seaborn
## First lets import seaborn 
import seaborn as sns

## What is seaborn? 
## Seaborn is a plotting/graphing package similar to matplotlib but has many astedically pleasing graphs and styles compared to matplotlib that many gravitate to
```

```python
## Lets try checking out working with a dataframe with seaborn!
## lets use pandas dataframe 

df = pd.DataFrame({'X':np.random.normal(0,2,1000),
                  'Y':np.random.normal(3,5,1000),
                  'Z':np.random.normal(-4,10,1000)})
```

```python
df.head()
```

```python
## Lets try making a KDP or kernal density plot with seaborn (similar to a histogram)
## notice how the histograms are plotted within the sns.kdeplot()

sns.set()
plt.figure(figsize=(6,4))
for col in df.columns:
    plt.hist(df[col],density=True, alpha=0.5,bins=30) ## First lets make a historgram from matplotlib to compare
    sns.kdeplot(df[col],fill=True,) ## the kde plot in sns
```

```python
## Now lets try a distrobution plot with seaborn

for col in df.columns:
    sns.distplot(df[col])
```

```python
## Now lets try setting a style within sns
## We'll also create a joint plot

with sns.axes_style('white'):
    sns.jointplot(data=df,x='X',y='Y',kind='kde')
```

```python
## Now lets make a pairplot from the iris dataset

sns.pairplot(iris,hue='species')
```

## Seaborn: Relplot

```python
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Today lets create some Relplots using seaborn!
## A relplot is a relationship plot which is a scatterplot to show a relationship between data!
## The data we're gonna use is the iris dataset so lets load that here

iris = sns.load_dataset('iris')
iris.head()
```

```python
## Now lets make the relplot for the sepal_length and sepal_width

sns.relplot(data=iris,x='sepal_length',y='sepal_width')
```

```python
## Now that we have the relplot lets try changing the color of the plots based on the species.
## That is where Hue comes into play!

sns.relplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
```

```python
## We can also change the style based on the species so their not all dots!

sns.relplot(data=iris,x='sepal_length',y='sepal_width',hue='species',style='species')
```

## Seaborn: Relplot Kind Line

```python
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## We will be using the iris data again!

iris = sns.load_dataset('iris')
iris.head()
```

```python
## Lets explore the relationship plot more!
## Lets create a line relationship plot for the following data

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line')
```

```python
## We can use ci=None to remove the shading around the line

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line',ci=None)
```

```python
## We can use ci='sd' for standard diviation to show the variation in the data

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line',ci='sd')
```

```python
## We can use the hue value again and break the data up by the species

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line',ci=None,hue='species')
```

```python
## We can change the style also to further distinguish the data

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line',ci=None,hue='species',style='species')
```

```python
## If we want the markers to be displayed on the lines

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line',ci=None,hue='species',style='species',dashes=True,markers=True)
```

```python
## If we wanted the lines to be the same but change up just the markers by the style we would set dashes to False

sns.relplot(data=iris,x='sepal_length',y='sepal_width',kind='line',ci=None,hue='species',style='species',dashes=False,markers=True)
```

## Seaborn: Relplot Facets

```python
import seaborn as sns
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Lets learn some facets of relplots
## First lets load the Tips dataset from seaborn

tips = sns.load_dataset('tips')
tips.head()
```

```python
## Lets view the relationship of total bill, tip and if the customer is a smoker between the lunch and dinner times

sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',style='smoker',col='time')
```

```python
## Lets try a different dataset now! The fmri dataset in seaborn

fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
## Lets see the relationship between the timepoint and the signal seperated by each subject?
## Lets set the colwrap to be 5 for spacing and we want line plots
## We are also setting the data to be regions that are 'frontal', which we can do by using the 'query' method to define that attribute
## Lets try it!

sns.relplot(data=fmri.query("region=='frontal'"),x='timepoint',y='signal',hue='event',style='event',col='subject',col_wrap=5, kind='line')
```

## Seaborn: Catplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Lets make a Catplot or category plot!
## First lets use the Tips data again!

tips = sns.load_dataset('tips')
tips.head()
```

```python
## alright now that we have the dataset loaded lets make the catplot!

sns.set(style='ticks',color_codes=True) ## Lets change up the style 

sns.catplot(data=tips,x='day',hue='day',y='total_bill')
```

```python
## Notice how there are a lot of the dots spaced out on the above
## Lets use jitter to make straight lines

sns.catplot(data=tips,x='day',y='total_bill',hue='day',jitter=False)
```

```python
## We can set the kinds of this catplot as well like if we use 'swarm'

sns.catplot(data=tips,x='day',y='total_bill',hue='day',kind='swarm')
```

```python
## We can also categorize the data using the hue

sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm')
```

```python
## We can change the kind to be boxplots 

sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box')
```

```python
## We can change the kind to be boxen plots

sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='boxen')
```

```python
## We can change the kind to be violin plots

sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='violin')
```

```python
## We can change the kind to be bar charts

sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='bar')
```

## Seaborn: Heatmaps

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Lets make some heatmaps! Very good with data science and matricies
## Lets first generate a dataframe

df = pd.DataFrame(np.random.rand(10,10),columns=list('ABCDEFGHIJ'))
df.head()
```

```python
## Now lets create the heatmap
## Lets address some of the parameters
## The annot is the annotation cells in the rows and columns
## The annot_kws is the annotation key words, kinda like sub parameters for the annotations
## In this example we set the size of the annotations

sns.heatmap(df,annot=True,annot_kws={'size':7})
```
