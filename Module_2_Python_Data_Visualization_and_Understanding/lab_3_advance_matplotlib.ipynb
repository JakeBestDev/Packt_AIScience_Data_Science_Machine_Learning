## Matplotlib: Multiple Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## Lets first generate a new plot and add some data


x = np.linspace(0,10,1000) ## Lets generate 1000 points from 0-10
y = np.sin(x) ## Let us now use np.sin() and input our x values

plt.style.use('fivethirtyeight') ## Lets try a new style
fig = plt.figure() ## Create a new figure, or activate an existing figure.
ax=plt.axes() ## Add an Axes to the current figure and make it the current Axes.

ax.plot(x,y) ## Now lets plot x and y on our figure ax
```

```python
## We can all create a similar plot like above using just the following:
## Notice how it saves the style 
## We use ax to save the Axes

plt.plot(x,y)
```

```python
## Now lets create a y1, y2 and y3 using np.cos and the formula y+y2-0.5
## y1 = y = np.sin(x)
## y2 = np.cos(x)
## y3 = y+y2-0.5

y1 = np.sin(x)
y2 = np.cos(x)
y3 = y+y2-0.5
```

```python
## Now lets plot y1,y2,y3
## Notice how we now have 3 different curves 
## Notice how it is difficult to tell which plot is which. We don't have a legend. We will go into detail on that in a later lesson


plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
## We can take the above and easily change the styles
## The style will be carried on through the whole notebook unless changed

plt.style.use('bmh')
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
## And again we can do the same thing using our ax variable and build a figure like above instead of using plt.plot()

fig = plt.figure()
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

## Matplotlib: Line Colors and Styles

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## You are able to set colors and line styles within the plt.plot() statement
## First lets set our x and y values
## We will have y1,y2,y3 

x = np.linspace(1,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## Now lets set up the figure

plt.style.use('_mpl-gallery') ## Lets use a new style
fig = plt.figure(figsize=(4,2)) ## Set the size of the figure 
ax = plt.axes() ## Lets establish the axes

ax.plot(x,y1,color='blue') ## Lets plot y1 with color blue
ax.plot(x,y2,color='g') ## Lets plot y2 with color 'g' or green to show a different way of calling a color
ax.plot(x,y3,color=(0.8,0.6,0.7)) ## Lets plot y3 with a specific color by defining the RGB color scales
```

```python
## Not only the colors we can set the linestyles

plt.style.use('seaborn-v0_8-bright') ## Lets use a new style
fig = plt.figure(figsize=(4,2)) ## Set the size of the figure 
ax = plt.axes() ## Lets establish the axes

ax.plot(x,y1,color='blue', linestyle='solid') ## Lets plot y1 with a solid line style
ax.plot(x,y2,color='g',linestyle='dashed') ## Lets plot y2 with a dashed line style
ax.plot(x,y3,color=(0.8,0.6,0.7),linestyle='dashdot') ## Lets plot y3 with a dotted dash line style
```

## Matplotlib: Plot Style Quiz
#### Make a plot with Matplotlib that:
#### 1) x = 0-20 with 1500 values
#### 2) y = cos(x)
#### 3) figure size is 6x4 
#### 4) Figure style is 'seaborn-v0_8-talk'
#### 5) The line color is orange
#### 6) the line style is dashed

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## 1) x = 0-20 with 1500 values

x = np.linspace(0,20,1500)
```

```python
## 2) y = cos(x)

y = np.cos(x)
```

```python
## 3) figure size is 6x4

fig = plt.figure(figsize=(6,4))
ax = plt.axes()
```

```python
## 4) Figure style is 'seaborn-v0_8-talk'

plt.style.use('seaborn-v0_8-talk')
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
```

```python
## 5) The line color is orange

fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.plot(x,y,color='orange')
```

```python
## 6) the line style is dashed

fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.plot(x,y,color='orange',linestyle='dashed')
```
