## Matplotlib: Multiple Plots

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## Lets first generate a new plot and add some data


x = np.linspace(0,10,1000) ## Lets generate 1000 points from 0-10
y = np.sin(x) ## Let us now use np.sin() and input our x values

plt.style.use('fivethirtyeight') ## Lets try a new style
fig = plt.figure() ## Create a new figure, or activate an existing figure.
ax=plt.axes() ## Add an Axes to the current figure and make it the current Axes.

ax.plot(x,y) ## Now lets plot x and y on our figure ax
```

```python
## We can all create a similar plot like above using just the following:
## Notice how it saves the style 
## We use ax to save the Axes

plt.plot(x,y)
```

```python
## Now lets create a y1, y2 and y3 using np.cos and the formula y+y2-0.5
## y1 = y = np.sin(x)
## y2 = np.cos(x)
## y3 = y+y2-0.5

y1 = np.sin(x)
y2 = np.cos(x)
y3 = y+y2-0.5
```

```python
## Now lets plot y1,y2,y3
## Notice how we now have 3 different curves 
## Notice how it is difficult to tell which plot is which. We don't have a legend. We will go into detail on that in a later lesson


plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
## We can take the above and easily change the styles
## The style will be carried on through the whole notebook unless changed

plt.style.use('bmh')
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
## And again we can do the same thing using our ax variable and build a figure like above instead of using plt.plot()

fig = plt.figure()
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## Lets define some variables for our plots
## Lets use the same x, y1,y2,y3 we have been using
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## Lets first show our original plots from eailier and how the line styles and colors we're defined

plt.style.use('seaborn-v0_8-bright') ## Lets use a new style
fig = plt.figure(figsize=(4,2)) ## Set the size of the figure 
ax = plt.axes() ## Lets establish the axes

ax.plot(x,y1,color='blue', linestyle='solid') ## Lets plot y1 with a solid line style
ax.plot(x,y2,color='g',linestyle='dashed') ## Lets plot y2 with a dashed line style
ax.plot(x,y3,color=(0.8,0.6,0.7),linestyle='dashdot') ## Lets plot y3 with a dotted dash line style
```

```python
## We can use shortcuts to define colors and types of lines on a plot
## Notice how we don't define the line style or color, we instead use shorthand like:
## '-b' which means a solid blue line like in y1

plt.style.use('fivethirtyeight') ## Lets set a new style
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.plot(x,y1,'-b') ## '-b' which means a solid blue line
ax.plot(x,y2,'--g') ## '--g' which means a dashed green line
ax.plot(x,y3,':r') ## ':r' which means dotted red line
```

## Matplotlib: Axis Limits

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## First lets define our x and y1,y2,y3
## We'll use the same ones as we have been

x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5
```

```python
## We can define axis limits the better view the data presented in a plot
## The axes and Axis is a little different where:
## Axes will be to the confines of the plot whereas
## Axis can be defined outside the plot confines
## Both of which can be used in tandem
## See the examples below

## Lets first show an axes plot

plt.style.use('seaborn-v0_8-pastel') ## Lets try a new style
plt.figure(figsize=(4,2),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r',linewidth=0.8) ## Lets create a red dash dot plot with a line width of 0.8
```

```python
## So now that we have our figure defined lets try defining some axes limits
## We can limit the x boundries using xlim() while also usinge axes function

plt.figure(figsize=(4,2),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r',linewidth=0.8) ## Lets create a red dash dot plot with a line width of 0.8

plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
```

```python
## We can do the same thing with y using ylim()

plt.figure(figsize=(4,2),dpi=90) ## Lets define our figure size. The dpi or dots per inch is how crisp the figure appears. The larger the dip the more crisp
ax = plt.axes()
ax.plot(x,y1,'-.r',linewidth=0.8) ## Lets create a red dash dot plot with a line width of 0.8

plt.xlim(-2,12) ## So now we see that the limit of x is now set from -2 to 12
plt.ylim(-1.5,1.5) ## So now we see that the limit of y is now set from -1.5 to 1.5
```

## Matplotlib: Set Function

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
## the .set() is primarily used to define annotations, axis, and other plot features for a defined plot
## So for example in the previous lessions we just used plt.xlabel() so for all plt plots the label defined will be present
## but what if we don't want that? What if we define a plot, like ax and bx, and want seperate xlabels.
## That is where the set() function comes into play. You can set for individual plots

## Lets give it a try but first define some variables
## Lets use the same ones we have been

x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2-0.5

## Lets also define some secondary variables

a = np.linspace(0,10,1000)
b1 = np.sin(a*2)
b2 = np.cos(a*2)
b3 = b1+b2-0.5
```

```python
## Lets create our ax plot first using the x and y1,y2,y3 variables
## Notice how I used 2 different variants of the set()
## There is a function set() to define multiple parameters
## There is also a set_xlim and set_ylim to define specifically those variables 

fig = plt.figure(figsize=(6,4),dpi=100)
ax = plt.axes()
ax.plot(x,y1,'--b',label='sin(x)')
ax.plot(x,y2,'-.g',label='cos(x)')
ax.plot(x,y3,':r',label='sin(x)-cos(x)-0.5')
ax.set(xlabel='X', ylabel='Y', title='ax X-Y Plot')
ax.set_xlim(-1,11)
ax.set_ylim(-2,2)
ax.legend(loc='upper right')
```

```python
## Noe lets define bx using the a, b1,b2,b3 variables
## Notice how it has its own labels and titles

fig = plt.figure(figsize=(6,4),dpi=100)
bx=plt.axes()
bx.plot(a,b1,'--b',label='sin(a*2)')
bx.plot(a,b2,'-.g',label='cos(a*2)')
bx.plot(a,b3,':r',label='sin(a*2)-cos(a*2)-0.5')
bx.set(xlabel='A',ylabel='B',title='bx A-B Plot')
bx.set_xlim(-1,11)
bx.set_ylim(-3,3)
bx.legend(loc='upper left')
```

## Matplotlib: Scatterplots

```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
## Today we will check out scatterplots using plt.scatter
## First lets create an x and y value. they will be random values
## We will also create a 'colors' data points at random to represent the different colors of the data points
## We will also create a 'sizes' data points at random to represent the different sizes of the data points
## The x,y,size, and color are all random but normally the data would be drawn from a dataframe/base of some kind to draw conclusions

x = np.random.rand(100)
y=np.random.rand(100)
colors = np.random.rand(100)
sizes = 150*np.random.rand(100)
```

```python
## Lets now create our scatter plot using plt.scatter()
## Notice how similar it is to plt.plot()

## We will assign our x and y as expected
## Our color will be assigned to the c parameter
## Our sizes will be assigned to the s parameter
## The alpha parameter is how see through the points are
## Then the cmap is the color map. With matplotlib there are a ton of colormaps availiable

## We have also created a colorbar using plt.colorbar(), essentially very similar to a legend based on the color shading and sizes

plt.figure(figsize=(6,4))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap=plt.cm.viridis)
plt.colorbar()
```

```python
## We can see how the alpha value changes if we increase it the points are less transparent

plt.figure(figsize=(6,4))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,cmap=plt.cm.viridis)
plt.colorbar()
```

```python
## We can see how the cmap changes if we change the colormap

plt.figure(figsize=(6,4))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,cmap=plt.cm.autumn)
plt.colorbar()
```

```python
## We can see how the cmap changes if we change the colormap

plt.figure(figsize=(6,4))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,cmap=plt.cm.PRGn)
plt.colorbar()
```

```python
## Lets create a contour plot
## First we need datapoints and we're gonna try something new 

## We will be using meshgrid in numpy, basically it creates all the possible combinations of the numbers given and makes a matrix
## So there are 2 matrixes, X and Y, 
## X has -1 and 1 combos
## Y has 5 and 8 combos

## When these two matrixes are combined onto a graph and assighed X and Y axes we will have:
## -1,5  1,5  -1,8  1,8 
## These are all of the possible computations for these two matrixes

X,Y = np.meshgrid([-1,1],[5,8])
```
