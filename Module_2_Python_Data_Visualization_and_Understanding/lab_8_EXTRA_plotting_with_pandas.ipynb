# EXTRA LAB: Plotting With Pandas

```python
import pandas as pd
import numpy as np
```

```python
## Lets create a dataframe and how how to plot using just pandas

df = pd.DataFrame({'A':np.random.normal(0,1,1000),
                  'B':np.random.normal(-2,2,1000),
                  'C':np.random.normal(5,4,1000),})
```

```python
df.head()
```

```python
## now lets just try plotting using plot()

df.plot()
```

```python
## Lets try plotting a histogram

df.plot.hist(alpha=0.5)
```

```python
## Lets change some of the properties of the hist

df.plot.hist(stacked=True,bins=30)
```

```python
## Lets create an area plot
## Due to the dataframe we created initially this areaplot is a little too busy so lets make a smaller one below

df.plot.area(stacked=False)
```

```python
## Lets try generating another dataframe!

df2 = pd.DataFrame(np.random.rand(10,4),columns=list('abcd'))
df2.head()
```

```python
## Now lets try a area plot again
## There we go much better!

df2.plot.area()
```

```python
## Lets try stacking with the area plot!

df2.plot.area(stacked=False)
```

```python
## Now lets create a new dataframe with matrixes and try some pie plots

df3 = pd.DataFrame(3*np.random.rand(4,2),index=list('abcd'),columns=list('xy'))
df3.head()
```

```python
## Now lets try a pie plot!

df3.plot.pie(subplots=True,figsize=(8,4))
```
