# Machine Learning Model From Scratch: Minimum-to-mean Distance Classifier

### The basis for the minimum distance to mean (MDM) classifier is the cluster centres. During classification, the Euclidean distances from a candidate feature vector to all the cluster centres are calculated. The candidate cell is assigned to the class that qualifies as the closest one.

```python
## First lets import some libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
## Now lets create some some synthetic data using make_blobs from sklearn

X,y = make_blobs(n_samples=200,centers=2)
```

```python
X.shape
```

```python
y.shape
```

```python
## Now lets check out the unique values of y
## Normally for minimum-to-mean or most classifiers we would use 2 integers. Some use 0,1 and others use -1,1. Lets change this to -1,1

np.unique(y)
```

```python
## Lets change to -1,1

y[y==0] = -1
```

```python
## now lets check y again

np.unique(y)
```

```python
## Now lets plot the data to see what it looks like

plt.scatter(X[:,0],X[:,1],c=y)
```

## Now to build the minimum classifier

```python
## First lets seperate the data into classes

idx = y==1
```

```python
X1 = X[idx,:]
```

```python
X1.shape
```

```python
X2 = X[~idx,:]
```

```python
X2.shape
```

```python
## Now lets compute the mean of X1

M1 = X1.mean(axis=0)
```

```python
M1
```

```python
## Now lets compute the mean of X2

M2 = X2.mean(axis=0)
```

```python
M2
```

```python
## Now lets pick some random data points from the data and try to classify it as -1 and 1

idx = np.random.randint(y.size)
r = X[idx,:]
```

```python
r
```

```python
## Now lets try to predict that point to determine if it belongs to class 1, X1 or class 2 X2
## we are calculating the Euclidean distance from M1 or mean 1 and going to compare to Euclidean distance of mean 2
## whichever it is closer to is how it is classified

d1 = ((r-M1)**2).sum() 
d2 = ((r-M2)**2).sum() 
```

```python
## Now lets create an if statement to determince which Euclidean distance is smaller
## If d1 is smaller than d2 then the datapoint is classified as a 1
## if d1 is larger than d2 then the datapoint is classified as -1

if d1 < d2:
    print(y[idx],1)
else:
    print(y[idx],-1)
```

```python
## Now lets put this together to generate random points and classifiy them 

idx = np.random.randint(y.size)
r = X[idx,:]
d1 = ((r-M1)**2).sum() 
d2 = ((r-M2)**2).sum() 

if d1 < d2:
    print(y[idx],1)
else:
    print(y[idx],-1)
```
