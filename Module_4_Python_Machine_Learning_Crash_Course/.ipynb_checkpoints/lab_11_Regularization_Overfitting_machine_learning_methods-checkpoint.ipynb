# Overfitting: Regularization


### Regularization is a technique in machine learning that helps prevent from overfitting. It works by introducing penalties term or constraints on the model's parameters during training. These penalties term encourage the model to avoid extreme or overly complex parameter values. By doing so, regularization prevents the model from fitting the training data too closely, which is a common cause of overfitting. Instead, it promotes a balance between model complexity and performance, leading to better generalization on new, unseen data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
## Lets make some synthetic data 

X = np.asarray([-np.random.rand(50),np.random.rand(50)])
X = X.flatten()
X = np.sort(X)
```

```python
X.shape
```

```python
## we now X, which is 50 negative and positive values for a total of 100 and their sorted
X
```

```python
## Now lets define y's function for its variables
## We will use X above and then define a new parameter series of a's. These a's are the coefficients

a1, a2, a3, a4, a5, a6 = 1000,10,10,1,100,100

y = a1*(X**5) + a2*(X**4) + a3*(X**3) + a4*(X**2) + a5*X + a6
```

```python
## Now lets plot X and y
## Notice how flexable the model is here 

plt.plot(X,y)
```

```python
## Now lets do it again but reduce the size of the coefficients to smaller values
## See how the polynomial isnt as flexable. This is because we reduced the sizes of the coefficients 

a1, a2, a3, a4, a5, a6 = 0.01,0.01,1,1,0.5,3

y = a1*(X**5) + a2*(X**4) + a3*(X**3) + a4*(X**2) + a5*X + a6

plt.plot(X,y)
```

## If the number of parameters are really small or if their magnitude/coefficient is small then your model will no longer be that flexible. And this precisely is called regularization.

## Regularization is helpful for almost every Overfitting issue
