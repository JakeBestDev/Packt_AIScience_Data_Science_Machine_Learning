# Machine Learning: Overfitting

### Overfitting is an undesirable machine learning behavior that occurs when the machine learning model gives accurate predictions for training data but not for new data. When data scientists use machine learning models for making predictions, they first train the model on a known data set.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

## For demonstrating Overfitting lets create some synthetic data and build regression models

```python
## First lets make the synthetic data

n = 20
X = np.arange(n)
y = 3*X-4+10*(np.random.randn(X.size))
```

```python
## Now lets view the data

plt.scatter(X,y)
```

```python
## Lets fit a line to this scatter using a polynomial model using the sklearn package. 
## This is normally what would be use for linear regression without building your own model

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
polyModel = PolynomialFeatures(degree=1)
X_poly = polyModel.fit_transform(X[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y)
y_pred = m.predict(X_poly)
```

```python
## Now lets plot and see how our regression model fits the data
## As we can see this fits fairly well but there are errors

plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
## Now lets change the degree of the PolynomialFeatures(degree=1) to be 2 and watch what happens to the graph.

polyModel = PolynomialFeatures(degree=2)
X_poly = polyModel.fit_transform(X[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y)
y_pred = m.predict(X_poly)
```

```python
## Notice how there is a curve now in the regression.
## The model is trying to fit the data based on its curve with a degree of 2.

plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
## There are still errors however so now lets try degree of 3

polyModel = PolynomialFeatures(degree=3)
X_poly = polyModel.fit_transform(X[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y)
y_pred = m.predict(X_poly)
```

```python
## Closer but still some error. Lets try 4 and onward

plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

```python
## At 10 we get fairly close and we can see that the model is getting closer to fitting the data accuratly

polyModel = PolynomialFeatures(degree=10)
X_poly = polyModel.fit_transform(X[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y)
y_pred = m.predict(X_poly)

plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

## Now lets try again with the same dataset but break it up into training and testing data.
## We will seperate our a certain number of X's and y's and use those to train and then see the model looks like compared to the entirity of X

```python
## First lets seperate the training data with X2 and y2
## Lets only use every 5th data point in X and y to train. Remeber there are 20 points total originally with n = 20

X2 = X[0:n:5]
y2 = y[0:n:5]
```

```python
X2
```

```python
y2
```

```python
## Now lets try generating our polynomial again with a degree of 4

polyModel = PolynomialFeatures(degree=4)
X_poly = polyModel.fit_transform(X2[:,np.newaxis])
m = LinearRegression()
m.fit(X_poly,y2)
y_pred = m.predict(X_poly)
```

```python
## Now lets view the graph
## We can see the model is very very accurate with the few training points it had. 

plt.scatter(X2,y2)
plt.plot(X2,y_pred,color='red')
```

```python
## Now lets test against the whole dataset
## Lets generate a new model 

X_poly_all = polyModel.fit_transform(X[:,np.newaxis])
y_pred = m.predict(X_poly_all)
```

```python
## See thats not great.
## The model fits well with the 3 trained datapoints but not the rest of the data

plt.scatter(X,y)
plt.plot(X,y_pred,color='red')
```

## The above is one of the main issues with machine learning models
## So what do you do to resolve this issue? 
## There is a golden ratio of degree's for the polynomial model that must be found that does not overfit or under fit the training data 
## A ratio that can then be applied to the unseen data really well
