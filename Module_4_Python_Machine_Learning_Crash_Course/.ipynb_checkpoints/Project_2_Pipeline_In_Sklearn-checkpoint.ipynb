# Project 2: Pipeline in Sklearn


### A machine learning pipeline is a series of interconnected data processing and modeling steps designed to automate, standardize and streamline the process of building, training, evaluating and deploying machine learning models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs

from sklearn.decomposition import PCA

sns.set()
```

## In this project, I'll be taking just an example of polynomial regression and I will be building a pipeline of polynomial features and linear regression inside.

```python
## lets import our new package for linear regression and polynomial features and pipeline

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
## Now lets define our function

## We will define our degree to be 2 initially which if you remember from previous lessons we use that for the flexability of the models
## We will also be taking defined key word arguments. which is the **kwargs as this can vary from function to function call

## So how this works is the pipeline will start with the PolynomialFeatures and then from that generate a LinearRegression and that will be the complete model
## If you do not call this function with the pipeline we will need to build the PolynomialFeatures and LinearRegression models each time

## We will essentially be building our pipeline and returning that pipeline

def Pol_Reg(degree=2,**kwargs):
    p = make_pipeline(PolynomialFeatures(degree),LinearRegression(**kwargs))
    return p
```

```python
## Now lets generate some dummy data and test our function

X = np.arange(30)
y = 3*(X**2)-10*X+34 ## Building our y vals but remember there is usually never a y function defined, just data points
```

```python
## Now lets fit our model and create our y_pred or y predict using our new function

y_pred = Pol_Reg(2).fit(X[:,np.newaxis],y).predict(X[:,np.newaxis])
```

```python
## Now lets plot and check out our new y_poly

plt.scatter(X,y)
plt.plot(X,y_pred,c='red')
```

## Now lets try with a X**3 polynomial with noise applied

```python
## Now lets try with a X**3 polynomial with noise applied

X = np.arange(30)
y = 3*(X**3)-10*(X**2)+34 + np.random.randn(X.size)
```

```python
## Now lets fit our model and create our y_pred or y predict using our new function

y_pred = Pol_Reg(2).fit(X[:,np.newaxis],y).predict(X[:,np.newaxis])
```

```python
## We can see now that the polynomial is a little off but still fairly accurate

plt.scatter(X,y)
plt.plot(X,y_pred,c='red')
```

## Now lets change the degrees to see if it fits better

```python
## Now lets try with a X**3 polynomial with noise applied

X = np.arange(30)
y = 3*(X**3)-10*(X**2)+34 + np.random.randn(X.size)
```

```python
## Now lets fit our model and create our y_pred or y predict using our new function

y_pred = Pol_Reg(3).fit(X[:,np.newaxis],y).predict(X[:,np.newaxis])
```

```python
## We can see now that the polynomial is very accurate again!
plt.scatter(X,y)
plt.plot(X,y_pred,c='red')
```
