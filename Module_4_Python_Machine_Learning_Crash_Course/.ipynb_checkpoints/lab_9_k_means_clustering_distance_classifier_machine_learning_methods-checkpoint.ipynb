# Machine Learning From Scratch: K-Means Clusterings

### K-means clustering is an unsupervised learning algorithm used for data clustering, which groups unlabeled data points into groups or clusters. It is one of the most popular clustering methods used in machine learning.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
## First lets make a synthetic dataset to create our k-means clustering model

X,y = make_blobs(n_samples=500,centers=3,cluster_std=1.7)
```

```python
## Now lets view the data with a scatter

plt.scatter(X[:,0],X[:,1])
```

## Now our goal is to find the mean of those 3 clusters. 
## Using those means we will then categorize all of the data into colors assocated with those means.
## This is how k-means works. We are grouping common data points based on their proxcimity to a common mean

```python
## Right now we can see we have 500 rows and 2 features. 
## We know above there are 3 clusters so ideally we'd want 3 features
## One for each of the above clusters 

X.shape
```

```python
## Lets now generate all indicies between 0 and 499 to find our means
## This is all of our data points randomly assigned from 0-499 and mixed up
## We will use this to generate 3 random means first to then lead us to our k-means values we want

r = np.random.permutation(np.arange(X.shape[0]))
r
```

```python
## Now we will pick 3 random variables and assign them as means. 
## The odds they are actually the means are so improbable that they are definatly wrong but this is where we will start
## Like I mentioned this is random since r had all values of X randomized 

M1 = X[r[0],:]
M2 = X[r[1],:]
M3 = X[r[2],:]
```

```python
## Lets view our means 

M1
```

```python
M2
```

```python
M3
```

```python
## Now we will define a function that takes the data from X and then assigns it to each of these means as groupings depending on which is closest

def makeGroups(X,M1,M2,M3):
    D1 = ((X-M1)**2).sum(axis=1) ## Lets calculate the distance of all the values of X from M1 and assign that as Distance 1
    D2 = ((X-M2)**2).sum(axis=1)
    D3 = ((X-M3)**2).sum(axis=1)

    D = np.hstack((D1[:,np.newaxis],D2[:,np.newaxis],D3[:,np.newaxis])) ## Now lets concatonate D1,D2,D3 into a horizontal stack. D is a 500x3 matrix with this

    y = np.argmin(D,axis=1) ## Now we will calculate the index of the record with the minimum distance in each feature

    return y


```

```python
y = makeGroups(X,M1,M2,M3)
```

```python
## Now lets view our y values

np.unique(y)
```

```python
## now lets check the means from the above compared to the color map
## Now we can see there is a bit of a mix with this
## Lets make this more acccurate and refine our k-means

plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn')
```

```python
## Now lets compute the means again from our data to refine it to categorize our data better using the first y we generated 

idx = y == 0
M1 = X[idx,:].mean(axis=0)
M2 = X[y==1,:].mean(axis=0)
M3 = X[y==2,:].mean(axis=0)
```

```python
## Now lets run our function again and graph or data

y = makeGroups(X,M1,M2,M3)
```

```python
## Much closer now! Still some issues at the bottom but the orange cluster has been fully defined 
## The red cluster is starting to take over the yellow cluster as the k-means model gets more accurate

plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn')
```

```python
## Lets gather our means again and run our function and see what happens

idx = y == 0
M1 = X[idx,:].mean(axis=0)
M2 = X[y==1,:].mean(axis=0)
M3 = X[y==2,:].mean(axis=0)
```

```python
## Now lets run our function again and graph or data

y = makeGroups(X,M1,M2,M3)
```

```python
## Wow! Thats much closer! Lets try once more!

plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn')
```

```python
## Lets gather our means again and run our function and see what happens

idx = y == 0
M1 = X[idx,:].mean(axis=0)
M2 = X[y==1,:].mean(axis=0)
M3 = X[y==2,:].mean(axis=0)
```

```python
## Now lets run our function again and graph or data

y = makeGroups(X,M1,M2,M3)
```

```python
## Fantastic! We have successfully categorized our datapoints based off a k-means model! 

plt.scatter(X[:,0],X[:,1],c=y,cmap='autumn')
```
