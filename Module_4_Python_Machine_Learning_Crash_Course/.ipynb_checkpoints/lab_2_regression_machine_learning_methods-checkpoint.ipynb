# Machine Learbing Methods: Regression

## Machine Learning Regression is a technique for investigating the relationship between independent variables or features and a dependent variable or outcome. It's used as a method for predictive modeling in machine learning, in which an algorithm is used to predict continuous outcomes.

## We will go more indepth in later labs linear regression, this is just the basics

```python
## First lets import the libraries we need

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## First lets try just creating some sample data and create a regression analysis for that data!

n = 20
X = np.arange(n) ## capital X as data. wherever the we use unsupervised learning or clustering,we will be using capital X
y = 3*(X**2) + 4*X - 100 ## small Y as targets. So basically just a line 
```

```python
## Now lets create a scatterplot mapping using X and y through matplotlib

plt.scatter(X,y)
```

### Lets learn a new resource!

## Now lets try creating a regression model using sklearn

```python
## First lets import the new library for creating a regression model
## Now you may be wondering "WHY import a linear model when the above data is not linear? Its curved?"
## The sklearn.linear_model is capable of analyzing the curve and extend to other types

from sklearn.linear_model import LinearRegression
```

```python
## First lets build a model. We will go more indepth later on this is just the basics of linear regression
## we use y for the model creation as it would normally be a group of variables like we printed above and using linear regression we want to create a static line to follow our curve or similar from above!

m = LinearRegression()
m.fit(X[:,np.newaxis],y) ## The newaxis adds a new dummy data axis to the data. We're doing this because X is one dimensional and we'd need a 2 dimensional array for it to work!
```

```python
## Now lets predict for y and create a new y_pred

y_pred = m.predict(X[:,np.newaxis])
```

```python
## Now lets plot the data using matplotlib!
## As you can see there is a basic trend line created that coorilates with the original curve. Lets see if we can't get some bend in the regression line we made

plt.scatter(X,y) ## original plot from above
plt.plot(X,y_pred,color='red') ## new plot using y_pred from the Linear Regression modeling. 
```

## Lets try to get the regression model we created more accurate and bend more to our curve!

```python
## We will go into this library in a further lession but for this example it allows us to more accuratly create our regression model!

from sklearn.preprocessing import PolynomialFeatures
```

```python
## Lets create our modeling and predicting again! Only this time impliment the PolynomialFeatures

## Lets impliment the new PolynomialFeatures by remodeling our X and y 
polyModel = PolynomialFeatures(degree=2) ## as mentioned we will go further into this later but we set the degree to 2
X_poly = polyModel.fit_transform(X[:,np.newaxis])

## Lets create the model
m = LinearRegression()
m.fit(X_poly,y) ## Notice how we're using the X_poly and not just X

## Next predict
y_Poly_pred = m.predict(X_poly)
```

```python
## Now lets plot again using matplotlib!
## Great! Now the new regression model fits our original model EXACTLY! It predicted the Y based off the X

plt.scatter(X,y) ## original plot from above
plt.plot(X,y_pred,color='red') ## new plot using y_pred from the Linear Regression modeling. 
plt.plot(X,y_Poly_pred,color='green') ## New poly plot we created 
```

## Now lets look at the coefficients!
#### Regression coefficients are the quantities by which the variables in a regression equation are multiplied.

```python
## Now lets look under the hood at the coefficents of the Regression model we created!
## Lets remember the original function we created for y 

## y = 3*(X**2) + 4*X - 100

## Look at the coefficients!
## The X**2 coefficient is 3, the X coefficient is 4,
## The intercept is 100 but set to 0 rather than -100 but this will be explained further

m.coef_
```
