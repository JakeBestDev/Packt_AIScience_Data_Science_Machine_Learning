# Machine Learning Methods: Clustering

### Clustering is an unsupervised machine learning technique designed to group unlabeled examples based on their similarity to each other.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Lets learn about clustering with synthetic data

```python
## First import the make_blobs from sklearn library to create the synthetic dataset

from sklearn.datasets import make_blobs
```

```python
## Now lets create the synthetic data similar to what we did in lab 1
## As mentioned previously we will go more indepth with sklearn library later but for right now the basics of the make_blobs are as follows:
## the n_samples are the number of variables
## The number of centers is the number of "gravity centers" you want the mix of data to be around. So defining 5 means there will be 5 clusters of the data present
## The cluster_std is how tightly packed the clusters of the synthetic data is per column by a matter of the defined std. A lower std is more closly packed data

X,y = make_blobs(n_samples=500,centers=5,cluster_std=0.6) 
```

```python
## now lets look at the data by plotting the data
## As you can see we have 5 clusters of data

plt.scatter(X[:,0],X[:,1])
```

```python
## Now lets view the shape of X
## As we can see we have 2 columns, one for the first and second feature of each dot's X and Y values that are being plotted

X.shape
```

## Now lets try a new party of sklearn! Kmeans from sklearn.cluser

```python
## As mentioned we will dive further into tge sklearn library and its tools but for right now this is to give basic understandings on clustering

from sklearn.cluster import KMeans
```

```python
## So lets define the kmeans by defining the number of clusters for our data

kmeans = KMeans(n_clusters=5)
```

```python
## Now using our synthetic X and y data we defined above lets build a model to give us better understanding of these clusters

kmeans.fit(X)
```

```python
## Now lets predict a y_pred like we had previously

y_pred = kmeans.predict(X)
```

```python
## Now lets create our plot! 
## As you can see the plotted points have been colored based off the above y_pred model and it accuratly colored those variables based on their clusters of data!

plt.scatter(X[:,0],X[:,1],c=y_pred,cmap='viridis')
```
