# Matplotlib Basics

## Matplotlib Intro

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Basic example of creating a line plots

x = np.linspace(0,10,1000)
y = np.sin(x)

plt.plot(x,y)
```

```python
## Basic example of creating a scatterplot

plt.scatter(x[::10],y[::10], color = 'red')
```

## Seaborn vs Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = np.linspace(0,10,1000)
y = np.cumsum(np.random.randn(1000,5),0)
```

```python
## Matplotlib Plotting

plt.style.use('classic')
plt.plot(x,y)
plt.legend('ABCDE',ncol=2,loc='upper left')
```

```python
## Seaborn Plotting

sns.set()
plt.plot(x,y)
plt.legend('ABCDE',ncol=2,loc='upper left')
```
