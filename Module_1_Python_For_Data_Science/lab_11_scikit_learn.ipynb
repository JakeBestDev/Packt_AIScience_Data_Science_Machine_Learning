## Scikit-Learn Linear Regression

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
## First lets set up some dummy data for our Linear Regression

N = 30
x = np.arange(N)
y = 4* x -7 +(5*np.random.randn(N))
plt.scatter(x,y)
```

```python
## Linear Regression Model

model = LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis],y) 
yfit = model.predict(x[:,np.newaxis])
plt.scatter(x,y)
plt.plot(x,yfit, color = 'red');
```

## Scikit-Learn make_blobs

```python
from sklearn.datasets._samples_generator import make_blobs
```

```python
## Creating datapoints using make_blobs

X,y = make_blobs(n_samples=500, centers=3,cluster_std=2)
plt.scatter(X[:,0], X[:,1], c=y, cmap='autumn')
```

## Scikit-learn train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
## Splitting data into training and testing data using train_test_split
## Split by test data = 33% and train data = 66% as indicated by the test_size = 0.33

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33)
```

## Scikit-learn SVC

```python
from sklearn.svm import SVC
```

```python
## creating a machine learning model classifier using smv or Support Vector Machine and the actual Classifier is the SVC

mdl = SVC(gamma='auto')
mdl.fit(X_train,y_train)
```

```python
## testing our classifier with our testing data 

y_predict = mdl.predict(X_test)
```

```python
## Determining the accuracy of the classifier 

acc = 100*np.sum(y_predict == y_test)/y_test.size
print('Accuracy is : ', acc,"%")
```

## Scikit-learn RandomForrestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
## RandomForrestClassifier is another classifier that will be covered more indepth in later classes

mdl = RandomForestClassifier(n_estimators=20)
mdl.fit(X_train,y_train)
```

```python
## Predicting the accuracy of the RandomForrestClassifier 

y_predict2 = mdl.predict(X_test)
acc2 = 100 * np.sum(y_predict2 == y_test)/y_test.size
print('Accuracy is : ', acc2,"%")
```
